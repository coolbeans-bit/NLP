```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import re
import requests
```

```python
# Download a sample text corpus
url = "https://www.gutenberg.org/files/1342/1342-0.txt"  # Pride and Prejudice
text = requests.get(url).text.lower()
```

```python
# Tokenize and clean text
words = re.findall(r'\b\w+\b', text)
word_counts = Counter(words)
```

```python
# Rank words by frequency
sorted_counts = sorted(word_counts.values(), reverse=True)
ranks = np.arange(1, len(sorted_counts) + 1)
frequencies = np.array(sorted_counts)
```

```python
# Zipf’s Law plot
plt.figure(figsize=(10, 5))
plt.loglog(ranks, frequencies, marker="o", linestyle="none", markersize=3, label="Observed")
plt.loglog(ranks, frequencies[0] / ranks, linestyle="--", color="red", label="Zipf’s Law (α≈1)")
plt.xlabel("Rank (r)")
plt.ylabel("Frequency f(r)")
plt.title("Zipf's Law in Text Corpus")
plt.legend()
plt.show()
```

```python
# Function to find number of words appearing exactly n times
def words_with_frequency_n(n):
    return sum(1 for freq in word_counts.values() if freq == n)

# Example: Number of words appearing exactly 5 times
n = 5
print(f"Number of words that appear exactly {n} times: {words_with_frequency_n(n)}")
```
